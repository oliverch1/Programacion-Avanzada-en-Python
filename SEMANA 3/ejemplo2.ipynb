```python
# Script de prueba
# Escriba un script que retorne la impresion de 12 registros en una tabla
#%%
from tabulate import tabulate
import random
import datetime

def datos_prueba(num_reg: int, hora_ini: str, minutos_offset: int) -> dict:

    idx = 1                                                                    # Inicio del índice
    formato_hora = "%H:%M"                                                     # Formato de la hora
    hora_actual = datetime.datetime.strptime(hora_ini, formato_hora)           # Conversión de la hora inicial a datetime
    for _ in range(num_reg):
       
        yield [idx, hora_actual.strftime(formato_hora), random.randint(50, 150), round(random.uniform(1.0, 2.0), 4)]
        hora_actual += datetime.timedelta(minutes=minutos_offset)
        idx += 1

gen = datos_prueba(num_reg=12, hora_ini="09:00", minutos_offset=30)            # Se genera los datos de prueba

datos = [next(gen) for _ in range(12)]

print(tabulate(datos, headers=["IDX", "HORA", "VAL1", "VAL2"], tablefmt="grid")) # Se imprimi una tabla con valores utilizando tabulate
# 
```

```python
# Resuelva el Script aqui
# -*- coding: utf-8 -*-
import json
import os
import random
from pathlib import Path
from dataclasses import dataclass

@dataclass
class Question:
    pregunta: str
    opc_1: str
    opc_2: str
    opc_3: str
    rspta: str

def load_questions(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        data = json.load(f)
    questions = [Question(**q) for q in data]
    return questions

def generate_test(questions, test_path, ballot_path):
    selected_questions = random.sample(questions, 10)
    with open(test_path, 'w', encoding='utf-8') as f:
        f.write('<!DOCTYPE html>\n<html lang="en">\n<head>\n')
        f.write('<meta charset="UTF-8">\n<meta name="viewport" content="width=device-width, initial-scale=1.0">\n')
        f.write('<meta http-equiv="X-UA-Compatible" content="ie=edge">\n<title>Test</title>\n</head>\n<body>\n')
        f.write('<main>\n<h1>Test</h1>\n<ol>\n')
        for i, q in enumerate(selected_questions, start=1):
            f.write(f'<li>{q.pregunta}</li>\n<ul style="list-style-type:none;">\n')
            f.write(f'<li>a. {q.opc_1}</li>\n')
            f.write(f'<li>b. {q.opc_2}</li>\n')
            f.write(f'<li>c. {q.opc_3}</li>\n')
            f.write('</ul>\n')
        f.write('</ol>\n</main>\n</body>\n</html>\n')
    with open(ballot_path, 'w', encoding='utf-8') as f:
        for i, q in enumerate(selected_questions, start=1):
            f.write(f'{i}. {q.rspta}\n')

def main():
    questions = load_questions('questions_bank.json')
    for i in range(1, 11):
        dir_path = Path.cwd() / f'Eval{i:02d}'
        dir_path.mkdir(parents=True, exist_ok=True)
        generate_test(questions, dir_path / 'test.html', dir_path / 'ballot.txt')

if __name__ == '__main__':
    main()
```
